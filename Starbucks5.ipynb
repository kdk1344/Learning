```python
# folium 을 이용한 지도 생성   (위도 37.570705 경도 126.981354 )

starbucks_map = folium.Map(location=[37.570705, 126.981354],tiles='Stamen Terrain', zoom_start=11)
starbucks_map
```

```python
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx,'위도']
    lng = seoul_starbucks.loc[idx,'경도']
    
    folium.CircleMarker(
    location=[lat,lng],
        fill=True, fill_color='red', 
        fill_opecity=1, color='yellow',
        weight=1,redius=2).add_to(starbucks_map)
    
starbucks_map
```
