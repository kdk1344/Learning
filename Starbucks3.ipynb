```python
# 시군구별 종사자수 사업체수

sgg_biz = pd.read_csv('report2.txt', sep='\t', header=2)
```

```python
 sgg_biz.head()
```

```python
condition = sgg_biz['동']=='소계'
sgg_biz_selected = sgg_biz[condition]
sgg_biz_selected.head()
```

```python
columns = ['자치구','사업체수','계']
sgg_biz_final = sgg_biz_selected[columns]
sgg_biz_final.head()
```

```python
sgg_biz_final.columns = ['시군구명','사업체수','종사자수']
sgg_biz_final.head()
```

```python
sgg_biz_final = sgg_biz_final.reset_index(drop=True)
sgg_biz_final.head()
```

```python
sgg_biz_final.to_excel('sgg_biz.xlsx', index=False)
```
